```python
import cv2

def resize_image(image, size=(224, 224)):
    return cv2.resize(image, size)
```

```python
#图像归一化处理
def normalize_image(image):
    return image / 255.0
```

```python
#高斯模糊
def apply_gaussian_blur(image):
    return cv2.GaussianBlur(image, (7, 7), 0)
```

```python
#增加噪声
def add_noise(image):
    noise = np.random.normal(0, 10, image.shape)
    noisy_image = np.clip(image + noise, 0, 255).astype(np.uint8)
    return noisy_image
```

```python
# 调整对比度和亮度
# alpha是对比度控制（1-3范围较好）
# beta是亮度控制
alpha = 1.5
beta = 50
#re_img = cv2.addWeighted(re_img, alpha, re_img, 1-alpha, beta)
```

```python
pre_x=[]
pre_y=[]
for i in range(len(data_list)):
    path = './dataset/Georgia/'+data_list[i]
    pre_y.append(get_Georgia_id(data_list[i]))
    img = cv2.imread(path)
    re_img=resize_image(img,(224,224))
    re_img=normalize_image(re_img)
    re_img=apply_gaussian_blur(re_img)
    re_img = cv2.addWeighted(re_img, alpha, re_img, 1-alpha, beta)
    pre_x.append(re_img.flatten())
pre_x = np.array(pre_x)
pre_y = np.array(pre_y)
```

```python
orl_pre_x=[]
orl_pre_y=[]
for i in range(len(data_list1)):
    path = './dataset/orl/'+data_list1[i]
    orl_pre_y.append(get_orl_id(data_list1[i]))
    img = cv2.imread(path)
    re_img=resize_image(img,(224,224))
    re_img=normalize_image(re_img)
    re_img=apply_gaussian_blur(re_img)
    orl_pre_x.append(re_img.flatten())
orl_pre_x = np.array(orl_pre_x)
orl_pre_y = np.array(orl_pre_y)
```

#[主成分分析(PCA)]+神经网络

```python
from sklearn.decomposition import PCA
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(pre_x,pre_y)
# PCA降维
pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(x_train)

# 构建神经网络模型
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train_pca.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(len(np.unique(y_train)), activation='softmax'))

# 编译模型
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 训练模型
model.fit(X_train_pca, y_train, epochs=300, batch_size=32)
```

```python
# 假设X_test是测试集特征
X_test_pca = pca.transform(x_test)  # 使用相同的PCA转换测试集

# 进行预测
predictions = model.predict(X_test_pca)

# 使用np.argmax获取最可能的类别索引
predicted_labels = np.argmax(predictions, axis=1)

# predicted_labels现在包含了模型预测的类别标签
print(predicted_labels)
```

```python
predicted_labels.shape
```

```python
from sklearn.metrics import accuracy_score

# 假设y_true是真实标签，y_pred是模型预测的标签
accuracy = accuracy_score(y_test, predicted_labels)
print(f"准确率: {accuracy}")
```

#[主成分分析(PCA)]+K近邻算法（KNN）（距离度量）

```python
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(pre_x, pre_y, test_size=0.3, random_state=42)

# 应用PCA进行降维
pca = PCA(n_components=0.95)  # 保留95%的方差
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# 使用KNN进行分类
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_pca, y_train)
predictions = knn.predict(X_test_pca)

# 打印准确率
accuracy = knn.score(X_test_pca, y_test)
print(f'Accuracy: {accuracy:.2f}')
```

#[主成分分析(PCA)]+支持向量机

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(pre_x, pre_y, test_size=0.3, random_state=42)

# 数据标准化
scaler = StandardScaler()

# 创建PCA转换器和SVM分类器
pca = PCA(n_components=0.95)  # 保留95%的方差
svm = SVC(kernel='linear', C=1.0)

# 创建管道，将PCA和SVM串联起来
pipeline = make_pipeline(scaler, pca, svm)

# 训练模型
pipeline.fit(X_train, y_train)

# 预测测试集
y_pred = pipeline.predict(X_test)

# 评估模型
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# 假设y_true是真实标签，y_pred是模型预测的标签
cm = confusion_matrix(y_test, y_pred)
person_names = []  # 举例：类别名称
for i in range(50):
    person_names.append(f'person{i}')

# 使用seaborn的heatmap函数可视化混淆矩阵
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=person_names, yticklabels=person_names)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# 展示正确分类的案例
print("\nCorrectly Classified Examples:")
for i in range(max(y_test)):
    correct_class_i = np.where(y_test == i)[0]  # 找到真实标签为i的索引
    correct_predicted_i = np.where(y_pred == i)[0]  # 找到预测标签为i的索引
    correct_indices = np.intersect1d(correct_class_i, correct_predicted_i)  # 找到两者相同的索引
    for idx in correct_indices:
        print(f"Example {idx + 1}: True - {person_names[y_test[idx]]}, Predicted - {person_names[y_pred[idx]]}")

# 展示错误分类的案例
print("\nIncorrectly Classified Examples:")
for i in range(max(y_test)):
    incorrect_class_i = np.where(y_test != i)[0]  # 找到真实标签不是i的索引
    incorrect_predicted_i = np.where(y_pred == i)[0]  # 找到预测标签为i的索引
    incorrect_indices = np.intersect1d(incorrect_class_i, incorrect_predicted_i)  # 找到两者相同的索引
    for idx in incorrect_indices:
        print(f"Example {idx + 1}: True - {person_names[y_test[idx]]}, Predicted - {person_names[y_pred[idx]]}")
```
